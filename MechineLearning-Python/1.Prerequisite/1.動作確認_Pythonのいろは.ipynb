# 関数

```
def multiply(a, b):
  return a * b

multiply(10, 20)
```

```
# 複数の値を返す関数
def mul_div(a, b):
  return a * b, a / b
m, d = mul_div(20, 10)
print(m, d)
```

# for文

```
for item in s:
  print(item)
```

```
for idx in range(len(s)):
  print(idx)
```

```
for idx,item in enumerate(s):
  print(idx, item)
```

# if文

```
if l1[0] == 1:
  print("l1[0] is 1")
```

```
if l1[0] == 0:
  print("l1[0] is 0")
else:
  print("l1[0] is not 1")
```

```
if l1[0] == 0:
  print("l1[0] is 0")
elif l1[0] == 1:
  print("l1[0] is 1")
```
